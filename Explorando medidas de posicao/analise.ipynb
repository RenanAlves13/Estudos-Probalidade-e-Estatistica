```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = r"C:/Users/Renanzin/.cache/kagglehub/datasets/sumedh1507/teen-phone-addiction/versions/1/"

print(os.listdir(path))
```

```python
dataset = pd.read_csv(f"{path}{os.listdir(path)[0]}")
```

```python
print(dataset.columns)
```

```python
dataset.head()
```

```python
colunas = ['Daily_Usage_Hours', 
                      'Sleep_Hours', 
                      'Academic_Performance',
                      'Social_Interactions', 
                      'Exercise_Hours', 
                      'Anxiety_Level',
                      'Depression_Level', 
                      'Self_Esteem', 
                      'Parental_Control',
                      'Screen_Time_Before_Bed', 
                      'Phone_Checks_Per_Day', 
                      'Apps_Used_Daily',
                      'Time_on_Social_Media', 
                      'Time_on_Gaming', 
                      'Time_on_Education', 
                      'Family_Communication', 
                      'Weekend_Usage_Hours',
                      'Addiction_Level']
```

```python
# Calculando a mediana de cada coluna
median_by_age = dataset.groupby('Age')[colunas].median()
print(median_by_age)
```

```python
media_by_age = dataset.groupby('Age')[colunas].mean()
```

```python
path_to_mean_png_file = "C:/Users/Renanzin/OneDrive/Documentos/Estatística básica/Explorando medidas de posicao/images/mean"
path_to_median_png_file = "C:/Users/Renanzin/OneDrive/Documentos/Estatística básica/Explorando medidas de posicao/images/median"
```

```python
for i in  colunas:
    media_by_column = media_by_age[i]

    media_by_column = pd.Series(media_by_column, name=i)
    media_by_column.index.name = 'Age'

    plt.figure(figsize=(7, 3)) 
    media_by_column.plot(kind='bar', color='skyblue')
    
    plt.title(f"{i} mean by age")
    plt.xlabel('Age')
    plt.ylabel(f"{i}")
    
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7) 
    
    plt.tight_layout()
    plt.savefig(f"{path_to_mean_png_file}/{i}", dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
for i in  colunas:
    media_by_column = media_by_age[i]

    media_by_column = pd.Series(media_by_column, name=i)
    media_by_column.index.name = 'Age'

    plt.figure(figsize=(7, 3)) 
    media_by_column.plot(kind='bar', color='skyblue')
    
    plt.title(f"{i} median by age")
    plt.xlabel('Age')
    plt.ylabel(f"{i}")
    
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7) 
    
    plt.tight_layout()
    plt.savefig(f"{path_to_median_png_file}/{i}", dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
# Calculando o desvio médio absoluto de cada variável
mediana = dataset[colunas].median()
print("Mediana para as colunas: \n", mediana)
```

```python
# Calculando o desvio médio absoluto de cada variável

print("\nMédia e Desvio Médio Absoluto para as colunas:\n")
media = dataset[colunas].mean()
mad = dataset[colunas].apply(lambda x: x.sub(x.mean()).abs().mean())

j = 0
for i in colunas:
    print(f"{i}\nMédia: {media[j]}{" "*10}DMA: {mad[j]}\n")
    j += 1
```

```python
# Calculando a variância

print("\nVariância para as colunas:\n")
variancia = dataset[colunas].var()
print(variancia)
```

```python
print("\nDesvio Padrão para as colunas:\n")
media = dataset[colunas].mean()
desvio_padrao = dataset[colunas].std()

j = 0
for i in colunas:
    print(f"{i}\nMédia: {media[j]}{" "*10}DP: {desvio_padrao[j]}\n")
    j += 1
```
